### Example 2

This example uses a single code cell and the slice() function.  The slice function returns a "slice object" that can be used inside the [] of a list slice.

```python
slice_013=slice(6,0,-2)
print( "line_013" )
print( line_013 )
print( f"{slice_013.start}:{slice_013.stop}:{slice_013.step}" )
print( line_013[slice_013] )
```
