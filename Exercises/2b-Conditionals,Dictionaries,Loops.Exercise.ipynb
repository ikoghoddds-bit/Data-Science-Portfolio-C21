1. Create a dictionary to store information about a famous person. Store their first name, last name, age, and what they are famous for. You should have keys ‘first_name’, ‘last_name’, ‘age’, and ‘famous_for’. Print each piece of information stored in your dictionary. **BONUS**: Use a for loop to print the information.

````python
famous_person = {
    "first_name": "Albert",
    "last_name" : "Einstein",
    "age"       : "147",
    "famous_for": ["Relativity", "Photoelectric Effect"],
}

for key, value in famous_person.items():
    print(f"{key}: {value}")
````

4. Write a python program that will take a word and print a message stating whether that word starts with a vowel or a consonant. Test it out with several words.

Example input:

```
popcorn
```

Example output:
```
This word starts with a consonant.
```


````python
word = "arable"

clean_word = word.lower() # Convert the word to lowercase so capitalization doesn't break our check

first_letter = clean_word[0] # Extract just the very first letter

if first_letter in "aeiou": # Check if that letter is inside our collection of vowels
    print("This word starts with a vowel.")
else:
    print("Word starts with a consonant.")
````

````python
word = input("Type your word here: ")

if word.lower()[0] in "aeiou":
    print("This word starts with a vowel.")
else:
    print("This word starts with a consonant.")
````
