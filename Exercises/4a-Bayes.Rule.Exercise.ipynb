## Problem 1:

Joan says she is itchy. There is a test for Allergy to Cats, but this test is not always
right:

- For people that really do have the allergy, the test says "Yes" 80% of the time
- For people that do not have the allergy, the test says "Yes" 10% of the time

If 1% of the population has the allergy, and Joan’s test says "Yes", what are the chances that Joan really has the allergy?


```python
# Probrior probabilities

Prob_Allergic = 0.01          # Prob(Allergy i.e 1%)
Prob_not_Allergic = 1 - Prob_Allergic   # Prob(No Allergy i.e. 99%)

# Conditional probabilities (Likelihoods)

Prob_Y_given_Allergic = 0.80      # Prob(Yes | Allergy)
Prob_Y_given_not_Allergic = 0.10  # Prob(Yes | No Allergy)

# Total probability of testing positive (Marginal Likelihood)


Prob_Y = (Prob_Y_given_Allergic * Prob_Allergic) + (Prob_Y_given_not_Allergic * Prob_not_Allergic)

# Probosterior probability using Bayes' Rule
Prob_A_given_Y = (Prob_Y_given_Allergic * Prob_Allergic) / Prob_Y

display(f"Total probability of testing 'Yes': {Prob_Y:.4f}")

display(f"Probability of having the allergy given a positive test: {Prob_A_given_Y:.4f} ({Prob_A_given_Y * 100:.2f}%)")
```
